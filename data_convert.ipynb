```python
# contains() 返回表示各str是否含有指定模式的字符串
# replace() 替换字符串
# lower() 小写
# upper() 大写
# split() 返回字符串中的单词列表
# strip() 删除前到和后置的空格
# join() 返回一个字符串
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('data/master.csv',encoding='utf-8')
df
```

```python
def f(x):
    if x == 'male':
        return '男'
    elif x == 'female':
        return '女'
    else:
        return '未知'

df['性别'] = df['sex'].apply(f)
df.head()
```

```python
df2 = df['sex'].map({'male':'男','female':'女'})
df2.head()
```

```python
df['year'] = df['year'].astype('str')
df3 = df['year'].apply(lambda x :str(x).replace(x[2:3],'**'))
df3
```
